```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 5,-2, 1]
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
x3 = np.random.normal(15, 5, 2000)
x4 = 4 + np.random.normal(loc=0, scale=2, size=60)
y4 = 4 + np.random.normal(loc=0, scale=2, size=len(x4))


ax[0,0].plot(y0)
ax[0,1].plot(x1, y1)
ax[0,2].hist(x3, bins=30)
ax[1,0].scatter(x4, y4)
plt.show()
```
